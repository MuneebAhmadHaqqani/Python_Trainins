### Ex1.

```python
def readfile(a):
    with open(a,"r") as f:
        a=f.read()
        return a
```

```python
print(readfile("new.txt"))
```

### Ex2.

```python
def read_append(a,b):
    f=open(a,"a")
    f.write(b)
    f.close()
    fr=open(a,"r")
    result=fr.read()
    fr.close()
    return result
```

```python
print(read_append("new.txt","\nnew line"))
```

```python
def readline(a):
    file=open(a,"r")
    result=file.readlines()
    return result
```

```python
print(readline("new.txt"))
```
