```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\etisalat\Downloads\Social_Network_Ads.csv")
```

```python
data.head()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

### split the data into x and y

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
x
```

```python
y
```

### using normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_sc=scaler.fit_transform(x)
```

```python
x_sc.max()
```

```python
x_sc.min()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x_sc,y,test_size=0.3)
```

### create KNN algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
model.score(xtest,ytest)
```

```python
model.predict_proba(xtest)
```

### visualization for the data

```python
plt.scatter(data[data["Purchased"]==1]["Age"],data[data["Purchased"]==1]["EstimatedSalary"])
plt.scatter(data[data["Purchased"]==0]["Age"],data[data["Purchased"]==0]["EstimatedSalary"])
plt.show()
```

### model evaluation using confusion matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(ytest,ypred)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,ypred))
```

### comparing the model with Logistic regression and selecting the best model based on confusion matrix

```python
from sklearn.linear_model import LogisticRegression
```

```python
model2=LogisticRegression()
```

```python
model2.fit(xtrain,ytrain)
```

```python
ypred2=model2.predict(xtest)
```

```python
print(classification_report(ytest,ypred2))
```

```python
confusion_matrix(ytest,ypred2)
```

### Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model3=DecisionTreeClassifier()
```

```python
model3.fit(xtrain,ytrain)
```

```python
ypred3=model3.predict(xtest)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
confusion_matrix(ytest,ypred3)
```

```python
print(classification_report(ytest,ypred3))
```
