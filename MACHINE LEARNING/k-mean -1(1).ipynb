```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
data_sc=scaler.fit_transform(data)
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=3)
```

```python
model.fit(data_sc)
```

```python
model.labels_
```

```python
data["cluster"]=model.labels_
```
