```python
import cv2
import matplotlib.pyplot as plt 
import matplotlib.image as im
import numpy as np
```

### load happy3.jpg image using (open cv)
### detect the face in the photo

```python
img=cv2.imread(r"C:\Users\etisalat\deep learning\kaizen_training\people.jpg")
```

```python
cv2.imshow("img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fd=cv2.CascadeClassifier(r"C:\Users\etisalat\deep learning\kaizen_training\kaizen\haarcascade_frontalface_alt.xml")
result=fd.detectMultiScale(img,1.03,4)
result
```

```python
for (x,y,w,h) in result:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),3)

cv2.imshow("img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### detect the face from the camera frame (single or multiple)

```python
fd=cv2.CascadeClassifier(r"C:\Users\etisalat\deep learning\kaizen_training\kaizen\haarcascade_frontalface_alt.xml")


vid=cv2.VideoCapture(0)

while True:
    ret,img2=vid.read()
    result=fd.detectMultiScale(img2,1.03,4)
    for (x,y,w,h) in result:
        cv2.rectangle(img2,(x,y),(x+w,y+h),(0,0,255),3)

    cv2.imshow("img",img2)

    if cv2.waitKey(1) & 0xFF==ord("q"):
        break
       
vid.release()
cv2.destroyAllWindows()
    
```
