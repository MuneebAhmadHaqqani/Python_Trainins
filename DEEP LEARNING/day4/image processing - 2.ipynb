### take  a frame from a camera

### load a video

```python
vid=cv2.VideoCapture(0) 

while True:
    res,img=vid.read() 
    cv2.imshow("image",img)

    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
vid.release()
cv2.destroyAllWindows() 
```

### convert red pixel to white

```python
# i is the row , j is the columns 
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img[i,j,0]>0.8 and img[i,j,1]<0.3 and img[i,j,2]<0.3:
            img[i,j,:]=1

            
plt.imshow(img)
```

### change one color from camera frame

```python
img_RGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
```

```python
vid=cv2.VideoCapture(0) 

while True:
    res,img=vid.read() 
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i,j,0]>100 and img[i,j,1]<50 and img[i,j,2]<50:
                img[i,j,:]=255 
                
    
    cv2.imshow("image",img)

    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
vid.release()
cv2.destroyAllWindows() 
```
