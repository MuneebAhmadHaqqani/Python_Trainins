```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as im
```

```python
vid=cv2.VideoCapture(0)  # 0 means default webcam

while True:
    res,img=vid.read()
    cv2.imshow("image",img)
    
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
    
vid.release()
cv2.destroyAllWindows()
```

```python
img_RGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(img_RGB)
plt.savefig("imgp.jpg")
```
