```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(123)
data=np.random.randint(10,100,(30,3))
data
```

```python
columns_names=["Temp_sensor","Hum_sensor","Power_sensor"]
```

```python
Dates=pd.date_range("07-01-2021",periods=30)
Dates
```

```python
table=pd.DataFrame(data,index=Dates,columns=columns_names)
table
```

### Extract based on a condition

### 1st way

```python
table[table["Temp_sensor"]>50]
```

```python
table[table["Temp_sensor"]>50]["Temp_sensor"]
```

```python
table[table["Temp_sensor"]>30][table["Hum_sensor"]>30]["Power_sensor"]["2021-07-20":"2021-07-25"]
```

### 2nd way --> logical operator

```python
table[(table["Temp_sensor"]>30) | (table["Hum_sensor"]>30)]
```

```python
table[(table["Temp_sensor"]>30) & (table["Hum_sensor"]>30)]
```
