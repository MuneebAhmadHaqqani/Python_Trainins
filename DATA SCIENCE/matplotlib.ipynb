```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib
```

### Line plot : using a list

```python
plt.figure(figsize=[10,10])

y1=[10000,5000,2000,3000,3500,4000,5000]
y2=[11000,6000,3000,4000,4500,5000,6000]
x=[1,2,3,4,5,6,7]

plt.plot(x,y1,marker="o",color="r",LineStyle="--",label="branch 1")
plt.plot(x,y2,marker="o",color="b",LineStyle="--",label="branch 2")

plt.title("sales figure")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()

plt.show()
```

### Example using a DataFrame 

```python
data=pd.DataFrame({"sales":[3000,5000,2000,2400,5000,1000,1500],
                    "Age":[20,30,40,35,45,55,60],
                   "Gender":[1,1,0,1,0,1,1],
                   "Salary":[20000,15000,30000,5000,25000,30000,22000]})
```

```python
data
```

```python

plt.plot(data["Age"],data["Salary"],marker="o",color="r",LineStyle="--")

plt.title("sales Vs. Age figure")
plt.xlabel("Age")
plt.ylabel("Sales")


plt.show()
```

### Showing two charts in one figure

```python

plt.subplot(1,2,1)

y1=[10000,5000,2000,3000,3500,4000,5000]
y2=[11000,6000,3000,4000,4500,5000,6000]
x=[1,2,3,4,5,6,7]

plt.plot(x,y1,marker="o",color="r",LineStyle="--",label="branch 1")
plt.plot(x,y2,marker="o",color="b",LineStyle="--",label="branch 2")

plt.title("sales figure")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()

plt.show()

plt.subplot(1,2,2)

plt.plot(data["Age"],data["Salary"],marker="o",color="r",LineStyle="--")

plt.title("sales Vs. Age figure")
plt.xlabel("Age")
plt.ylabel("Sales")


plt.show()
```

### Bar Chart

```python

y1=[10000,5000,2000,3000,3500,4000,5000]
x=[1,2,3,4,5,6,7]

plt.bar(x,y1,color="r")
plt.plot(x,y1,marker="o",color="black",LineStyle="--",label="branch 1")

plt.title("sales figure")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()

plt.show()
```

### Scatter plot

```python
plt.scatter(data["Age"],data["Salary"],c="r",marker="o",s=90)
plt.show()
```

### PieChart

```python
per=[70,30]
label=["V","Not V"]
c=["g","r"]


plt.pie(per,labels=label,colors=c,startangle=100,explode=(0,0.05),autopct="%1.0f%%")
plt.title("V Vs. Not V.")

plt.show()
```

### Creating mutiple charts in one graph

```python
plt.subplot(2,2,1)
plt.plot(data["Age"],data["Salary"],marker="o",color="r",LineStyle="--")

plt.title("sales Vs. Age figure")
plt.xlabel("Age")
plt.ylabel("Sales")


plt.show()

plt.subplot(2,2,2)
gender=["M","M","F","M","M","F","M"]

plt.hist(gender)
plt.show()

plt.subplot(2,2,3)
plt.scatter(data["Age"],data["Salary"],c="r",marker="o",s=90)
plt.show()

plt.subplot(2,2,4)
per=[70,30]
label=["V","Not V"]
c=["g","r"]


plt.pie(per,labels=label,colors=c,startangle=100,explode=(0,0.05),autopct="%1.0f%%")
plt.title("V Vs. Not V.")

plt.show()

plt.savefig("Multiple Charts")
```
